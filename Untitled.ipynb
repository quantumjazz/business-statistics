## Линейни уравнения

Линейните уравнения и техните решения са фундаментални в алгебрата, като полагат основите за разбиране на връзки между променливи величини. Разбирането на тези връзки е нашата основна задача.

## Теорема 1: Свойства на равенствата

Ще получим еквивалентно уравнение, ако:

1. Едно и също количество се добавя или изважда от всяка страна на дадено уравнение.
2. Всяка страна на дадено уравнение се умножава или дели на едно и също ненулево количество.

## Пример 1:

**Решете и проверете отговора:**

$8x - 3(x - 4) = 3(x - 4) + 6$

#### Решение

- използваме дистрибутивното свойство: $8x - 3x + 12 = 3x - 12 + 6$
- разместваме: $5x + 12 = 3x - 6$
- изваждаме $3x$ от двете страни: $2x + 12 = -6$
- изваждаме $12$ от двете страни: $2x = -18$
- разделяме двете страни на $2$: $x = -9$

#### Проверка:

Заместваме $x = -9$ обратно в оригиналното уравнение:

$8(-9) - 3[(-9) - 4] = 3[(-9) - 4] + 6$

## Линейни неравенства

Разбирането на неравенствата е от съществено значение за решаването на проблеми, които изискват променливите в тях да бъдат изразени като интервали, а не като конкретни стойности.

### Пример 5: Решаване на линейно неравенство

**Решете:** $2(2x + 3) \leq 6(x - 2) + 10$

**Стъпки на решение:**

1. **Разпределяме и опростяваме:** $4x + 6 \leq 6x - 12 + 10$
2. **Комбинираме:** $4x + 6 \leq 6x - 2$
3. **Изолираме променливата $x$:** $-2x \leq -8 \Rightarrow x \geq 4$

| Дефиниране | нотация |            |
|-------------------|---------------------|------------|
| `[a, b]`          | `a ≤ x ≤ b`         | затворен интервал|
| `[a, b)`          | `a ≤ x < b`         | полуотворен интервал |
| `(a, b]`          | `a < x ≤ b`         | полуотворен интервал|
| `(a, b)`          | `a < x < b`         | отворен интервал|
| `(-∞, a]`         | `x ≤ a`             |   |
| `(-∞, a)`         | `x < a`             |   |
| `[b, ∞)`          | `x ≥ b`             |   |
| `(b, ∞)`          | `x > b`             |   |

- с `[` и `]` отбелязваме затворени интервали, включително крайните точки.
- с `(` и `)` отбелязваме интервали, които не вклчват крайните точки.
- между `-∞` и `∞` е интервалът на реалните числа.

```python
import matplotlib.pyplot as plt

# Create the plot
fig, ax = plt.subplots()

# Set the limits of the plot
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# Draw the axes
ax.axhline(y=0, color='black', linewidth=1)
ax.axvline(x=0, color='black', linewidth=1)

# Label quadrants
quadrant_labels = ['I', 'II', 'III', 'IV']
quadrant_coords = [(5, 5), (-5, 5), (-5, -5), (5, -5)]
for label, (x, y) in zip(quadrant_labels, quadrant_coords):
    ax.text(x, y, label, fontsize=12, ha='center', va='center')

# Mark a specific point (e.g., (3, 2)) and label it
specific_point = (3, 2)
ax.plot(specific_point[0], specific_point[1], 'ro')  # Red dot
ax.text(specific_point[0], specific_point[1], f' ({specific_point[0]}, {specific_point[1]})\n(Abscissa, Ordinate)',
        verticalalignment='bottom', horizontalalignment='left')

# Set labels
ax.set_xlabel('X-axis (Abscissa)')
ax.set_ylabel('Y-axis (Ordinate)')

# Show the plot
plt.show()
```
