## Данни

- Данните са резултати от измерване на характеристики на определени обекти или процеси. 

- Когато тези измервания имат различни числови или нечислови (количествени или качествени) стойности, ги наричаме *променливи*.

- Някои характеристики не се променят и ги наричаме *константи*.

## Примери

- *Променливи*: пол, възраст, успех, продажби, борсов индекс

- *Константи*: $\pi = 3,141$, $\exp = 2,718$, студенти от I курс, когато те са единственото множество, което наблюдаваме.

## Данни

- *Числови* – интервални и абсолютни. Могат да се подреждат по-големина и разстоянието между тях е измеримо;
- *Категориални (или символни)* – категорните данни нямат свойствата на числовите. Те могат да са еднакви или да се различават като стойност.
- Примери: цвят на очи, град, пол, държава, видове цветя, име на фирма, ден от седмицата, месец от годината.

## Категорални данни - ординални

- Ординални данни (ordinal data) – специален под тип на категорните данни. При този тип ние не знаем, разстоянията между отделните 
стойности, но все пак може да ги подредим.
- Пример 1: променлива наречена „магнитуд“, която може да приема стойности - висок, среден и нисък.
- Пример 2: Месеците в годината. Знаем кой след кой следва, но точните разстояния са относителни.

## Категорални данни - номинални 

- Номинални (nominal) – втория подтип на категорните данни,
уникалните стойности на подобен тип данни не се сравними по никакъв
начин.
- Пример: Колона от данни с имена на фирми.
- Преобразуване: Преди използване на категорни променливи, те често
се преобразуват в числа. Стойностите на ординалните променливи
могат да се заменят с последователни числа. 

## Източници на данните

Винаги използваме данни от най-представителните източници, до които имаме достъп. Винаги посочваме източниците на данните.

- Примери:
- Макроикономически данни за България - от НСИ, Евростат, БНБ или Световната Банка;
- Фирмени данни от Търговския Регистър, специализирани бази данни.
- Като цяло избягваме да използваме данни от сайтове, чиито източници не са известни.

## Формати на данните

- Два най-често срещани: "широк" и "дълъг".
- В широкия формат на данни всеки ред представлява един обект (напр. човек), а всяка колона съдържа различни променливи или наблюдения за този обект в различно време или при различни условия.
- В дългия формат всеки ред представлява едно наблюдение или измерване, често включващо времето или състоянието на наблюдението. Това може да направи набора от данни „по-дълъг“ в сравнение с широкия формат.

## Широк Формат

| PersonID | Measurement_1        | Measurement_2        | Measurement_3        |
|----------|----------------------|----------------------|----------------------|
| 1        | 70kg                 | 71kg                 | 70kg                 |
| 2        | 80kg                 | 82kg                 | 81kg                 |
| 3        | 60kg                 | 59kg                 | 60kg                 |

Всяка колона представлява различно измерване за една и съща променлива (тегло).

## Широк Формат

- Подходящ за ситуации, в които имате относително малък брой наблюдения.
- Често се използва в описателната статистика, където фокусът е върху сравняването на променливите за всеки индивид.
- Може да стане тромав ако броят на точките за наблюдение, условията или променливите се увеличи значително.
- Може да доведе до излишък и неефективност при съхранение и манипулиране на данни, особено ако много стойности липсват.

| PersonID | Measurement          | Weight |
|----------|----------------------|--------|
| 1        | 1                    | 70kg   |
| 1        | 2                    | 71kg   |
| 1        | 3                    | 70kg   |
| 2        | 1                    | 80kg   |
| 2        | 2                    | 82kg   |
| 2        | 3                    | 81kg   |
| 3        | 1                    | 60kg   |
| 3        | 2                    | 59kg   |
| 3        | 3                    | 60kg   |

Всеки ред представлява различно измерване, което улеснява проследяването на промените във времето или при различните условия.

## Дълъг Формат

- Подходящ за данни с множество наблюдения във времето или при различни условия за едни и същи субекти.     
- Предпочитан за анализ на времеви серии и други данни с повтарящи се измервания.
- Полезен в ситуации, при които наборът от данни може да се разшири с повече наблюдения или променливи с течение на времето.
- Може да бъде по-малко интуитивен за директно тълкуване, тъй като данните на всеки субект са разпределени в множество редове.     

## Процес на статистическа обработка на данните

Разграничават се следните стъпки:
1. Установяване на проблема и формулировка на хипотезите;
2. Събиране на данни;
3. Обработка на събраните данни;
4. Конструиране на модела;
5. Интерпретиране на модела и установяване на изводите.

## Обработка на данните {.scrollable}

- Включва в себе си поне две обичайни задачи:
- Проверка за *изключителни стойности* (Outliers) - това са стойности, които
не са съвместими с останалите такива в дадено множество от данни
(например твърде големи са или твърде малки). Изключителните
стойности често са грешки в измерванията на приборите или човешки
такива или аномалии в субектите/процесите, които са наблюдавани. При установяване на изключителни стойности следва те да бъдат
премахнати или третирани със специфични техники;
- Скалиране, преобразуване и избиране на характеристики - например, ако
една характеристика е в метри, а другата е в сантиметри, то те следва да се
преобразуват в една мерна единица.
- Повече за това - в следваща лекция.
